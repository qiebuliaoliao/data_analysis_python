# 主要内容——4.2 通用函数：快速点对点数组函数

ufunc,是用来在数组中实现点对点操作的，可以理解为把一些简单的函数快速的向量化封装，输入和输出都是一个以上的标量。很多ufuncs都是点对点变换，例如sqrt和exp

```python
import numpy as np
arr = np.arange(10)
arr
```

```python
np.sqrt(arr)
```

```python
np.exp(arr)
```

这种函数叫做一元通用函数(unary ufuncs)，其他例如add, maximum的函数，需要两个数组作为结果

```python
x = np.random.randn(8)
y = np.random.randn(8)
x
```

```python
y
```

```python
np.maximum(x,y)
```

这里是点对点的比较x和y中的元素

尽管不常见，ufunc也能返回多个数组，如modf，返回小数和整数部分

```python
arr = np.random.randn(7)*5
arr
```

```python
a, b = np.modf(arr)
a
```

```python
b
```

ufun可以接收一个可选参数作为输出，这样可以直接更改原有的数组。例如：

```python
arr
```

```python
np.sqrt(arr)
```

```python
arr
```

这个操作并没有改变原来的数组，但是以下这个操作会改变数组

```python
np.sqrt(arr, arr)
```

```python
arr
```
